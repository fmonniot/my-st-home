```python
def f(x):
    return 1.3340933096745218e+000 + 2.1058530062491476e-002 * x + (7.7819337068634094e-003 * (x ** 2))

plt.plot(xdata, ydata, label = "truth")
plt.plot(xdata, [ f(x) for x in xdata ], label = "approximation")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.show()
```
